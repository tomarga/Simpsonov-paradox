## 1 Uvod <a name="uvod"></a>

*Simpsonov paradoks* (ili Yule-Simpsonov učinak) fenomen je u vjerojatnosti i statistici, u kojem se trend pojavljuje u nekoliko različitih skupina podataka, ali nestaje ili je obrnut kada se isti podaci agregiraju. Ovaj se rezultat često susreće u društvenim i medicinsko-znanstvenim statističkim analizama i posebno je problematičan kada se podacima o frekvenciji neke pojave neoprezno dodjeljuju krive interpretacije. Paradoks se može riješiti kada se u statističkom modeliranju prikladno vodi računa o uzročnim odnosima. 
Edward H. Simpson prvi je opisao ovaj fenomen u tehničkom članku 1951. godine.

## 3 Simpsonov paradoks <a name="paradoks"></a>

### 3.2 Definicija

Nakon uvodnog primjera dajemo definiciju problema i to prvo za slučaj binarne slučajne varijable.

**Definicija 1.** (Binarni slučaj)<br>
Promatramo $n$ skupina podataka takvih da svaka skupina $i$ ima $A_i$ pokušaja i $\ 0 \leq a_i\leq A_i$ uspjeha. Slično, za skupinu $i$ vrijedi da od $B_i$ pokušaja ima $0\leq b_i\leq B_i$ uspjeha. Kažemo da se Simpsonov paradoks pojavio ako vrijedi sljedeće:  $$\frac{a_i}{A_i}\leq\frac{b_i}{B_i}\ \forall i=1,2,3,...,n, \ \ ali \ \ \frac{\sum_{i=1}^{n}{a_i}}{\sum_{i=1}^{n}{A_i}}\leq \frac{\sum_{i=1}^{n}{b_i}}{\sum_{i=1}^{n}{B_i}}.$$

###  3.1 Vektorski prikaz

Jednostavan slučaj Simpsonovog paradoksa se može ilustrirati pomoću dvodimenzionalnog vektorskog prostora. U binarnom slučaju razina uspješnosti se prikazuje kao omjer uspjeha u odnosu na broj pokušaja ($\frac{p}{q}$). Navedenu mjeru uspješnosti možemo prikazati kao dvodimenzionalni vektor $\vec{A}=(q,p)$. Strmiji vektor tada predstavlja veću stopu uspjeha. Ako su dvije stope $\frac{p_1}{q_1}$ i $\frac{p_2}{q_2}$ povezane, rezultat može biti reprezentiran kao suma   
vektora $(q_1,p_1)+(q_2,p_2)$ koja prema pravilu paralelograma ima nagib $\frac{p_1+p_2}{q_1+q_2}$. 

Simpsonov paradoks nalaže da iako vektor $\vec{a_1}$ ima manji nagib nego vektor $\vec{b_1}$, te analogno $\vec{a_2}$ ima manji nagib nego vektor $\vec{b_2}$, suma vektora $\vec{a_1} + \vec{a_2}$ može potencijano imati veći nagib od sume $\vec{b_1} + \vec{b_2}$. Navedeno je prikazano u donjem primjeru, gdje su vektori s manjim nagibom prikazani rozom, a vektori s većim nagibom, žutom bojom. 

```python
fig1, ax1 = plt.subplots()
ax1.set_xlim( 0, 2.5 )
ax1.set_ylim( 0, 2.5 )
ax1.set_title( 'Vektorski prikaz' )

a1 = ( [1,0 ])
a2 = ( [3,3] )
a_rez = ( a1[0] + a2[0], a1[1] + a2[1] )
b1 = ( [3,1] )
b2 = ( [0.5,1] )
b_rez = ( b1[0] + b2[0], b1[1] + b2[1] )

origin = [0.5], [0.5]
plt.quiver( *origin, a1[0], a1[1], scale=10, color = 'hotpink', label = '$a_1$' )
plt.quiver( *origin, a2[0], a2[1], scale=10, color = 'hotpink', label = '$a_2$' )
plt.quiver( *origin, b1[0], b1[1], scale=10, color = 'gold', label = '$b_1$' )
plt.quiver( *origin, b2[0], b2[1], scale=10, color = 'gold', label = '$b_2$' )
plt.quiver( *origin, a_rez[0], a_rez[1], scale=10, color = 'purple', label = '$a_1+a_2$' )
plt.quiver( *origin, b_rez[0], b_rez[1], scale=10, color = 'orange', label = '$b_1+b_2$' )
plt.legend()

plt.show()
```

### 3.2 Obrat korelacije (*Correlation reversal* )

Moguće je definirati Simpsonov paradoks za kontinuirane varijable. Jednostavan primjer bio bi pogled na kovarijancu $\sigma(x,y)$ između dvije varijable, koja je različitog predznaka za subpopulacije te za cijelu populaciju. Kovarijanca cijele populacije je dana s dvije komponente: težinska suma kovarijanci subpopulacija te kovarijanca očekivanja subpopulacije.

Obrat korelacije jasnije se vidi na sljedećim grafičkim prikazima. Podaci korišteni za prikaze su umjetno generirani po uzoru na [2] te nemaju neku opipljivu interpretaciju. Podatke iz subpopulacije i kovarijance koriste funkciju nasumičnog odabira pa je moguće da nećemo uvijek dobiti uzorak koji pokazuje paradoks. Svejedno, u prosjeku dobivamo zadovoljavajuć uzorak.

```python
#f-ja prima broj podataka koje generiramo
#uzorak dijeli u dvije subpopulacije
def generiraj_uzorak( n_samples ):

    overall_cov = 3*np.array( [[1,0.9], [0.9,1]] )

    means = np.random.multivariate_normal( mean=[0,0], cov=overall_cov, size=2 )
    
    weights = np.random.uniform( size=2 )
    weights /= np.sum( weights )
    covs = [np.random.uniform( 0.2,0.8 ) for _ in range(2)]
    covs = [np.array( [[1,-c], [-c,1]] ) for c in covs]

    samples = []

    for sg, ( mean, cov, w ) in enumerate( zip( means, covs, weights ) ):
        n = int( round( n_samples * w ) )
        sample = np.random.multivariate_normal( mean=mean, cov=cov, size=n )
        sample = pd.DataFrame( sample, columns=["x", "y"] )
        sample["z"] = sg
        samples.append( sample )
        
    df = pd.concat( samples )
    
    return df
```

```python
df = generiraj_uzorak( 300 )

ukupna_kov = ( df[["x", "y"]].cov().iloc[0,1] )
prva_populacija_kov = df[df.z==0][["x", "y"]].cov().iloc[0,1]
druga_populacija_kov = df[df.z==1][["x", "y"]].cov().iloc[0,1]

print( 'Ukupna kovarijanca: ', ukupna_kov )
print( 'Kovarijanca prve subpopulacije', prva_populacija_kov )
print( 'Kovarijanca druge subpopulacije', druga_populacija_kov )
```

Već vidimo da smo dobili dobar uzorak - kovarijance ukupne populacije i subpopulacija su suprotnog predznaka.

```python
fig, ax = plt.subplots( figsize=( 10,5 ) )
ax.set_title( 'Korelacija agregiranih podataka', fontsize=15 )
sns.regplot( data=df, x="x", y="y" )
plt.show()
```

```python
fig, ax = plt.subplots( figsize = ( 10, 5 ) )
ax.set_title( 'Korelacija grupiranih podataka', fontsize=15 )

#regplot za prvu grupu
sns.regplot( data=df[df.z==0], x="x", y="y", ax=ax )
#regplot za drugu grupu
sns.regplot( data=df[df.z==1], x="x", y="y", ax=ax )
plt.show()
```

Po nagibu pravaca očito je da je za ukupnu populaciju korelacija $x$ i $y$ pozitivna, a za svaku od subpopulacija korelacija je negativna.

## 4 Studija slučaja diskriminacije<a name="case_study"></a>

### 4.1 Skup podataka

Slijedi analiza konkretnog slučaja etničke diskriminacije pri dodjeljivanju sredstava osobama zaostalim u razvoju. Koristimo podatke sa stranice _Kaggle_ [1]. Podaci se odnose na slučaj u Kaliforniji, iz 2014. godine. 

Prikupljeni su podaci za 300 osoba. Za svaku osobu znamo kojeg je spola, kojoj etničkoj i dobnoj skupini pripada te koliku pomoć u obliku novčanih sredstva prima. 

```python
url = 'https://raw.githubusercontent.com/tomarga/Sympsonov-paradox/master/californiaDDSDataV2.csv'
df = pd.read_csv( url, index_col=0 )
df.head()
```

Ovdje nećemo analizirati podatke vezane uz spol primatelja pomoći tako da iste brišemo iz tablice. 

```python
df.drop( [ 'Age', 'Gender'], axis=1, inplace=True )
# brisanje podataka o skupinama s premalo podataka za analiziranje
df = df[df.Ethnicity != 'Other']
df = df[df.Ethnicity != 'Native Hawaiian']
```

Budući da nas zanima postoji li diskriminacija po etničkoj pripadnosti, promotrimo prvo etničke skupine koje analiziramo.

```python
ethnicity = set( df['Ethnicity'] )
ethnicity = np.array( ethnicity )

population = df.groupby( 'Ethnicity' )['Age Cohort'].count()

df.Ethnicity.value_counts().plot.pie( figsize=( 6, 6 ), autopct='%1.1f%%', pctdistance=0.7, shadow = True )
plt.show()
```

Vidimo da baratamo s podacima o osobama iz 6 različitih etničkih skupina. Najviše podataka se odnosi na populaciju španjolskog podrijetla te bijelu populaciju nešpanjolskog podrijetla. 

### 4.2 Analiza podataka

Analizu započinjemo usporedbom ukupne količine novčane pomoći u odnosu na različite etničke skupine. Navedeno prikazujemo  odgovarajućim dijagramima pravokutnika.

```python
colors=["c","c","c","m","c","r"]
figure, ax = plt.subplots( figsize=( 15,6 ) ) 
bplot=sns.boxplot( x='Ethnicity', y='Expenditures',data=df.sort_values( 'Ethnicity' ), width= 0.6 , palette=colors )

bplot.axes.set_title( 'Količina sredstava po etničkim skupinama', fontsize=15 ) 
bplot.set_xlabel( "Etnička skupina", fontsize=10 )
bplot.set_ylabel( "Količina sredstava", fontsize=10 ) 
bplot.tick_params( labelsize=10 )
```

Uočavamo najveći nesrazmjer između novčane pomoći prema _Hispanic_ i _Multi Race_ skupini(niska ulaganja) te _White not Hispanic_ skupini(visoka ulaganja). Budući da za _Hispanic_ skupinu postoji puno više podataka, usredotočit ćemo se na njenu daljnu usporednu s _White not Hispanic_ skupinom. Navedene skupine su radi isticanja obojane drugim bojama. Istaknimo još ovdje kako iako joj je medijan dosta nizak, _Hispanic_ skupina podataka sadrži popriličan broj _outlier_ -a. Još ne možemo brzati sa zakljičcima o diskriminaciji, uzimajući u obzir da pokazujemo samo sumu novčanih sredstava, a ne prosječnu količinu sredstava koju pojedina osoba određene etničke skupine zapravo prima.

Zato nadalje promatramo prosjek zaprimljene novčane pomoći.

```python
prosjecna = df.groupby( 'Ethnicity' ).mean().sort_values( 'Expenditures' )
prosjecna.head( 6 )
```

Već iz tablice je očito da je i prosjek primanja značajno niži za _Hispanic_ skupinu. Svejedno, radi lakše usporedbe prosjeke prikazujemo u obliku stupčanih dijagrama.

```python
dict = prosjecna.to_dict()['Expenditures']

dict = sorted( dict.items() )
keys = [elem[0] for elem in dict]
values = [elem[1] for elem in dict]

figure, ax = plt.subplots(figsize=(15,5)) 
ax.set_title( 'Prosječna količina sredstava po etničkim skupinama', fontsize=15 )
plt.bar( keys, values, color='cccmcr' )
plt.xlabel( "Etnička skupina", fontsize=10 )
plt.ylabel( "Količina sredstava", fontsize=10 ) 
plt.show()
```

Ljubičasti( _Hispanic_ ) stupac je znatno niži od crvenog( _White not Hispanic_ ) stupca. Dakle, prosječna pomoć koju prima osoba s poteškoćama u razvoju španjolskog podrijetla je manja od one koje dobiva bijela osoba nešpanjolskog podrijetla. Je li stvarno prisutna diskriminacija prema osobama španjolskog podrijetla ili nešto propuštamo? Pokušajmo naći moguća objašnjenja u slučaju da nije riječ o diskriminaciji. Mogli bismo tako pretpostaviti da su osobe španjolskog podrijetla u Kaliforniji manje upoznate s mogućnostima prijave za novčanu pomoć ili možda imaju veću potporu s obiteljske strane.  

U dosadašnjoj analizi smo zanemarili dobnu skupinu primatelja naknade. Sada uspoređujemo prosječnu količinu dobivenih sredstava dviju promatranih etničkih skupina, ali unutar istih dobnih skupina.

```python
white = df[df['Ethnicity'] == 'White not Hispanic']
white = white.groupby( 'Age Cohort' ).mean().sort_values( 'Expenditures' )

hispanic = df[df['Ethnicity'] == 'Hispanic']
hispanic = hispanic.groupby( 'Age Cohort' ).mean().sort_values( 'Expenditures' )

age_white_hisp = white.merge( hispanic,left_index=True, right_index=True )
age_white_hisp.columns = ['White not Hispanic', 'Hispanic']
age_white_hisp.index.names = ['Dobna skupina']
age_white_hisp.head( 6 )
```

```python
age_white_hisp.plot( kind="bar", figsize=( 15,5 ), color='rm', 
                           title = "Prosječna količina sredstava po etničkim skupinama i starosti",
                           rot=0 ).title.set_size( 15 )
plt.ylabel( "Količina sredstava", fontsize = 10 )
plt.show()
```

Gornji _Hispanic_ i _White not Hispanic_ stupci su u istim dobnim skupinama gotovi jednaki! Štoviše, u svim dobnim skupinam osim jedne(18-21), prosječna količina potpore za _Hispanic_ skupinu je viša od suprostavljene. Da sumiramo - u određenim manjim grupama podataka, za svaku grupu, _Hispanic_ skupina u prosjeku dobiva višu potporu, dok u sveukupnoj populaciji _White not Hispanic_ skupina u prosjeku dobiva značajno višu potporu. Dakle, svjedočimo Simpsonovom paradoksu! Varijabla dobne skupine se u ovom primjeru pokazala kao _lurking_ varijabla, bez koje bi bili došli do krivog zaključka. Diskriminacije po gornjem prikazu nema. Zašto se pojavljuju dvije suprostavljene interpretacije? Pogledajmo detaljnije podatke s kojima baratamo.   

```python
white = df[df['Ethnicity'] == 'White not Hispanic']
white = white.groupby( 'Age Cohort' ).count()
white.drop( 'Expenditures',axis=1, inplace=True )

hispanic = df[df['Ethnicity'] == 'Hispanic']
hispanic = hispanic.groupby( 'Age Cohort' ).count()
hispanic.drop( 'Expenditures',axis=1, inplace=True )

age_white_hisp = white.merge( hispanic,left_index=True, right_index=True )
age_white_hisp.columns = ['White not Hispanic', 'Hispanic']
age_white_hisp.index.names = ['Dobna skupina']
age_white_hisp
```

```python
age_white_hisp.plot( kind="bar", figsize=(15,5), color='rm', 
                           title = "Broj primatelja pomoći etničkim skupinama i starosti",
                           rot=0 ).title.set_size( 15 )
plt.ylabel( "Broj primatelji pomoći", fontsize = 10 )
plt.show()
```

Količina podataka za ove dvije skupine su gotovo iste pa je ova usporedba relevantna. Broj primatelja pomoći iz _Hispanic_ skupine je veći u nižim dobnim skupinama, dok je broj primatelja pomoći iz _White not Hispanic_ skupine veći u starijim skupinama. Iz ovih podataka možemo razumijeti dobiveni paradoks. Naime, sveukupna _Hispanic_ populacija koja prima pomoć je mlađa od _White not Hispanic_ populacije koja prima pomoć. Budući da starije osobe dobivaju veću potporu, analiza agregiranih podataka rezultira prividom o diskriminaciji. 

## 5 Zaključak<a name="zakljucak"></a>

Simpsonov paradoks podrazumijeva određeni trend koji se pojavljuje u nekoliko različitih skupina podataka, ali nestaje ili je obrnut kada se isti podaci agregiraju. Uslijed istog, pri početnoj analizi moguće je intuitivno krivo zaključiti o povezanosti određenih značajki. Paradoks je relativno lako razriješiti detaljnijom analizom koreliranosti varijabli te detektiranjem _lurking_ varijabli. Zbog toga se uvijek dobro podsjetiti da podatke treba analizirati detaljno te promatrati iz različitih perspektiva.
